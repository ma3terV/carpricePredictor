```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('car data.csv')
print(df.shape)
print(df.columns)
df
```

```python
y=df.Selling_Price
df.drop(columns=['Selling_Price','Car_Name'], axis=1, inplace=True)
df
```

```python
df.replace({'Manual':0,'Automatic':1,'Diesel':0, 'Petrol':1, 'CNG':2,'Individual':0, 'Dealer':1},inplace=True)
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lg=LinearRegression()
x=df.values
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.1)
lg.fit(xtrain,ytrain)
print("R2_SCORE ON Training :",r2_score(ytrain,lg.predict(xtrain)))
print("R2_SCORE ON TESTING :",r2_score(ytest,lg.predict(xtest)))
```
